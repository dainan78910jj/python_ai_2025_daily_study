```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./messy_data.csv')
print(df.head(12))
```

```python
nan_data = df[df.isnull().any(axis=1)]
print(nan_data)
```

```python
df['age'] = df['age'].fillna(30)
df['height'] = df['height'].fillna(df['height'].mean().round(1))
df['income'] = df['income'].fillna(0)
df['purchases'] = df['purchases'].fillna(',,')
df
```

```python
unrealistic_ages = df[(df['age'] > 100) | (df['age']< 0)]
unrealistic_ages
```

```python
df['age'] = df['age'].mask(df['age'] > 100, 100)
df['age'] = df['age'].mask(df['age'] < 0, 0)
df
```

```python
unrealistic_heights = df[(df['height'] > 220) | (df['height'] < 100)]
unrealistic_heights
```

```python
df['height'] = df['height'].mask(df['height'] < 0, df['height'].mean())
df
```

```python
df['gender'] = df['gender'].replace('male', '0001').replace('female', '0010').replace('unknown', '0100').replace('other', '1000')
print(df)
```

```python
def insert_zero(x):
    x = x.split(',')
    
    for i in range(len(x)):
        if x[i] == '':
            x[i] = '0'
    
    x = ','.join(x)
    return x

df['purchases'] = df['purchases'].apply(insert_zero)
df
```

```python
df.to_csv('./cleaned_data.csv')
```
