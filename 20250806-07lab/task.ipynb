```python
import pandas as pd
```

```python
data_of_females = pd.read_csv('./female.csv')
data_of_males = pd.read_csv('./male.csv')
```

```python
print(f'For women we have (rows, columns) {data_of_females.shape}')
print(f'For men we have (rows, columns) {data_of_males.shape}')
```

```python
def compute_percentile_ranges(column):
    # Define percentile ranges
    ranges = [(0, 5), (5, 25), (25, 50), (50, 75), (75,90), (90, 97), (97, 100)]

    percentiles = {(low, high): (column.quantile(low/100), column.quantile(high/100)) for low, high in ranges}
    
    print(percentiles)

    counts = {}

    for r, (low, high) in percentiles.items():
        counts[r] = ((column >= low) & (column < high)).sum()
    
    return counts

print(compute_percentile_ranges(data_of_females['chestcircumference']))
print(compute_percentile_ranges(data_of_females['biacromialbreadth']))

print(compute_percentile_ranges(data_of_males['chestcircumference']))
print(compute_percentile_ranges(data_of_males['biacromialbreadth']))
```

```python
def compute_size_percentile_measurements(data, chest_column, shoulder_column):
    sizes = ['XS', 'S', 'M', 'L', 'XL', '2XL', '3XL']
    ranges = [5, 25, 50, 75, 90, 97, 100]

    #Compute the values for each percentile for chest and shoulder
    chest_percentiles = {}
    shoulder_percentiles = {}
    for p in ranges:
        chest_percentiles[p] = data[chest_column].quantile(p/100)
        shoulder_percentiles[p] = data[shoulder_column].quantile(p/100)
    
    # Map the t-shirt sizes to the corresponding chest and shoulder measurements
    size_mappings = {}
    for i, size in enumerate(sizes):
        size_mappings[size] = {
            'Chest': int(chest_percentiles[ranges[i]]),
            'Shoulder': int(shoulder_percentiles[ranges[i]])
        }
    
    return size_mappings

female_size_chart = compute_size_percentile_measurements(data_of_females, 'chestcircumference', 'biacromialbreadth')
male_size_chart = compute_size_percentile_measurements(data_of_males, 'chestcircumference', 'biacromialbreadth')

print(female_size_chart)
print(male_size_chart)
```

Last time, we created a function get_size to get a clearer view of how many matches and ties we had. Now, I want you to do the same thing for the new size charts we created, but this time taking into consideration that we have two measurements instead of one. The goal remains the same: find out how many matches and how many ties we have.
 
Task
Analyze the data: Use the new size charts to determine the number of matches and ties based on two measurements.
Count matches and ties: Write a function that iterates through each person's measurements, compares them with the new size charts, and counts the number of matches and ties.

Bonus
Modify the function to handle ties. If there is a tie and the sizes are adjacent, choose the larger size to increase the number of matches.

```python
def get_size(measurements, size_chart):
    
    res = {
        'Chest':'',
        'Shoulder':''
    }
    
    Chest = measurements['chestcircumference']
    Shoulder = measurements['biacromialbreadth']
    
    for size in size_chart:
        if Chest < size_chart[size]['Chest']:
            if res['Chest'] == '' or size_chart[size]['Chest'] < size_chart[res['Chest']]['Chest']:
                res['Chest'] = size
            
            
    for size in size_chart:
        if Shoulder < size_chart[size]['Shoulder']:
            if res['Shoulder'] == '' or size_chart[size]['Shoulder'] < size_chart[res['Shoulder']]['Shoulder']:
                res['Shoulder'] = size
    
    return res

measurement_1 = {
    'chestcircumference': 920,
    'biacromialbreadth': 360
}


measurement_2 = {
    'chestcircumference': 920,
    'biacromialbreadth': 400
}

print(get_size(measurement_1, female_size_chart))
print(get_size(measurement_2, female_size_chart))
```

```python
def calc_size_for_male(m):
    return get_size(m, male_sizes)

def calc_size_for_female(m):
    return get_size(m, female_sizes)

size_of_males = data_of_males.apply(calc_size_for_male, axis=1)

size_of_females = data_of_females.apply(calc_size_for_female, axis=1)

size_of_males
size_of_females
```

```python
def sum(input):

    def match(size):
        if size['Chest'] == size['Shoulder']:
            return True
        else:
            return False
        
    res = input.apply(match)

    sum_of_matches = 0
    for i in res:
        if i == True:
            sum_of_matches += 1
    print(f'We have {sum_of_matches} recrords that match.')


    sum_of_ties = 0
    for i in res:
        if i == False:
            sum_of_ties += 1
            
    print(f'We have {sum_of_ties} recrords that do not match.')
    
sum(size_of_females)
sum(size_of_males)
```
