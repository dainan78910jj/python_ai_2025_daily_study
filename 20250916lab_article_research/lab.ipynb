```python
content_1 = open("./articles/article_1.txt")

result_1 = content_1.read()

print(result_1)
```

```python
from openai import OpenAI
```

```python
ollama = OpenAI(base_url='http://host.docker.internal:11434/v1', api_key='')

model_name = 'llama3.2'

messages=[
    {"role": "system", "content": "Summarize this article in 3 bullet points, with brief sentences."},
    {"role": "user", "content": result_1}
]

response = ollama.chat.completions.create(
    model=model_name,
    messages=messages)

answer = response.choices[0].message.content
answer
```

```python
# define a variable 'summaries' to store the summary for all the articles
# for loop files in articles folder
#    open one text file and read in content, save to a local variable 'content'
#    put the content to the messages
#    put messages to API ollama.chat.completions.create as the value of the second param, get the response
#    save the response.choices[0].message.content to summaries
```

```python
import os
from research import save_summary

summaries = []

for filename in os.listdir("./articles/"):
    
    content = open(os.path.join("./articles/", filename)).read()
    
    messages=[
        {"role": "system", "content": "Summarize this article in 3 bullet points, with brief sentences."},
        {"role": "user", "content": content}
    ]
    
    response = ollama.chat.completions.create(
        model=model_name,
        messages=messages
    )
    
    text = response.choices[0].message.content
    
    summaries.append(text)

    saved_path = save_summary(text, summaries_dir="./summaries", prefix=filename)
    print("saved: ", saved_path)
```

```python
messages=[
    {"role": "system", "content": "Compare these articles. What are the common themes and key differences?"},
    {"role": "user", "content": '\n'.join(summaries)}
]

response = ollama.chat.completions.create(
    model=model_name,
    messages=messages)

answer = response.choices[0].message.content
answer
```
