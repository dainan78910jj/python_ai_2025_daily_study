1, What is the difference between stopping and deleting a VM?


Stopping a VM is a shutdown of the VM. It allows the VM to save its state and continue running after a start later. Deleting a VM is a permanent deletion of the VM and its data from disk.

2, How is a public IP used for accessing the VM?


A public IP address is used to access a virtual machine (VM) in a cloud computing environment over public network. It is a unique identifier that allows devices outside the cloud to communicate with the VM.


3, What are some real-world scenarios where VMs are used?


Virtual machines (VMs) are used in applications that sensitive to performance, latency, or security, or may need to be run on dedicated hardware.



4, How can you secure your VM against unauthorized access?

1. Use strong passwords and password policies.
2. Restrict access to the VM by using a firewall.
3. Force to use secured communication protocols.
4. Monitor your VM for suspicious activity.
5. Keep your operating system and software up to date.
